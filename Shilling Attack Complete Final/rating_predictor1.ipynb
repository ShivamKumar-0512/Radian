```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv("movie_data.csv", header=0)
data.head(5)
```

```python
label_encoder = LabelEncoder()
data['Title_encoded'] = label_encoder.fit_transform(data['Title'])
data['Director_encoded'] = label_encoder.fit_transform(data['Director'])
data['Actors_encoded'] = label_encoder.fit_transform(data['Actors'])
data['Genres_encoded'] = label_encoder.fit_transform(data['Genres'])
data['Duration_encoded'] = label_encoder.fit_transform(data['Duration'])

# print(data['Title'])
# print(data['Director_encoded'])

# Features
X = data[['Title_encoded','Director_encoded', 'Actors_encoded', 'Duration_encoded', 'Genres_encoded']]  
y = data['Rating'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.01, random_state=42)
```

```python
knn =  KNeighborsRegressor(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)


mse = mean_squared_error(y_test, y_pred)
rmse = sqrt(mse)
# Root Mean Square Error

print("RMSE--> ", rmse)
```
